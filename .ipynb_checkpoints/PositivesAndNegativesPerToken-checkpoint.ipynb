```python
df_text = df_text.astype(str)
```

```python
df_text['answer'] = df_text['answer'].apply(lambda x: x.lower())
```

```python
import string
df_text['answer']=df_text['answer'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
df_text.head()
```

```python
inverted_index = {}
```

```python
for _,row in df_text.iterrows():
    for token in row['answer'].split():
        print(token)
        if token in inverted_index:
            ##add this image too to the list of dict[token]
            inverted_index[token].append(row['image_url'])
        else:
            new_list = [row['image_url']]
            inverted_index[token]=new_list
```

```python
inverted_index
```

```python
## FIND ALL UNIQUE TOKENS
unique_tokens = []
for _,row in df_text.iterrows():
    for token in row['answer'].split():
        if token not in unique_tokens:
            unique_tokens.append(token)
```
