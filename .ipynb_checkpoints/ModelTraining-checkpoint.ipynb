```python
from sklearn.utils import shuffle

def get_training_data(token):
  X_train_token = []
  Y_train_token = []
  count = 0
  for url in positive_data[token]:
  #    print(url)
      if count<50:
        #print('Loading pos',url)

        img_feature = np.load('/content/drive/My Drive/data-ML/features_crop_extract/features_crop/' + url.split('/')[1].replace('.png','.npy'),allow_pickle=True)
        #print(img_feature)
        reshaped_img = img_feature.reshape((img_feature.shape[0]*img_feature.shape[1]), img_feature.shape[2])
        new_img = reshaped_img.transpose()
        X_train_token.append(new_img) 
        Y_train_token.append(1)
        count+=1

  count=0
  for url in negatve_data[token]:
    if count<50:
      #print('Loading neg',url)
      img_feature = np.load('/content/drive/My Drive/data-ML/features_crop_extract/features_crop/' + url.split('/')[1].replace('.png','.npy'))
      #print(img_feature)
      reshaped_img = img_feature.reshape((img_feature.shape[0]*img_feature.shape[1]), img_feature.shape[2])
      new_img = reshaped_img.transpose()
      X_train_token.append(new_img) 
      Y_train_token.append(-1)
      count+=1

  X_token,y_token = shuffle(X_train_token, Y_train_token)

  X_train_token = np.array(X_token)
  Y_train_token = np.array(y_token)
  nsamples, nx, ny = X_train_token.shape
  train_dataset = X_train_token.reshape((nsamples,nx*ny))

  return train_dataset,Y_train_token
```

```python
##CREATE ARRAY OF CLASSIFIERS
classifiers=[]
from sklearn.linear_model import LogisticRegression

for token in tokens:
  print(token)
  logisticRegr = LogisticRegression()
  train_dataset,labels = get_training_data(token)
  logisticRegr.fit(train_dataset,labels)
  classifiers.append(logisticRegr)
```

```python
with open('/content/drive/My Drive/data-ML/classifiers_26.pkl', 'wb') as f:
  pickle.dump(classifiers,f)
```

```python
from sklearn import svm
##CREATE ARRAY OF CLASSIFIERS
classifiers_svm=[]
for token in tokens[0:100]:
  clf = svm.SVC()
  print(token)
  train_dataset,labels = get_training_data(token)
  clf.fit(train_dataset,labels)
  classifiers_svm.append(clf)
```

```python
with open('/content/drive/My Drive/data-ML/svm_classifiers.pkl', 'wb') as f:
  pickle.dump(classifiers_svm,f)
```
