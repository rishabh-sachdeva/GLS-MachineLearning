## Fetch Image Summarization Textual Data collected Via MTurk

```python
df_text = pd.read_csv('/home/rishabh/Desktop/project/mturk-audio-data-analysis/textual_data/mturk_text.csv')
```

```python
df_text.head()
```

```python
df_text.iloc[0].image1_url.split('/')[len(df_text.iloc[0].image1_url.split('/'))-2]+'/'+df_text.iloc[0].image1_url.split('/')[len(df_text.iloc[0].image1_url.split('/'))-1]
```

### Simplify column values - keep instance name only

```python
df_text.image1_url = df_text.image1_url.map(lambda name : name.split('/')[len(name.split('/'))-2]+'/'+name.split('/')[len(name.split('/'))-1])
df_text.image2_url = df_text.image2_url.map(lambda name : name.split('/')[len(name.split('/'))-2]+'/'+name.split('/')[len(name.split('/'))-1])
df_text.image3_url = df_text.image3_url.map(lambda name : name.split('/')[len(name.split('/'))-2]+'/'+name.split('/')[len(name.split('/'))-1])
df_text.image4_url = df_text.image4_url.map(lambda name : name.split('/')[len(name.split('/'))-2]+'/'+name.split('/')[len(name.split('/'))-1])
df_text.image5_url = df_text.image5_url.map(lambda name : name.split('/')[len(name.split('/'))-2]+'/'+name.split('/')[len(name.split('/'))-1])
```

```python
df_text
```

```python
df_text_1 = df_text.loc[:,['image1_url','answer1']]
df_text_1=df_text_1.rename(columns={"image1_url":"image_url","answer1":"answer"})

df_text_2 = df_text.loc[:,['image2_url','answer2']]
df_text_2=df_text_2.rename(columns={"image2_url":"image_url","answer2":"answer"})

df_text_3 = df_text.loc[:,['image3_url','answer3']]
df_text_3=df_text_3.rename(columns={"image3_url":"image_url","answer3":"answer"})

df_text_4 = df_text.loc[:,['image4_url','answer4']]
df_text_4=df_text_4.rename(columns={"image4_url":"image_url","answer4":"answer"})

df_text_5 = df_text.loc[:,['image5_url','answer5']]
df_text_5=df_text_5.rename(columns={"image5_url":"image_url","answer5":"answer"})
```

```python
combinined_text_frames = [df_text_1,df_text_2,df_text_3,df_text_4,df_text_5]
df_text_com = pd.concat(combinined_text_frames)
```
